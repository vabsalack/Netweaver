# Metrics Monitoring

- The library automatically configures the appropriate matplotlib backend to `matplotlib.use("module://ipympl.backend_nbagg")` ensure live and interactive rendering of training metrics. By running matplotlib in a separate process and thread, the plots remain responsive and update in real time. For more details on the implementation, refer to the library's source code.

- Import the `PlotTraining` class from `netweaver.utils`.  

- Instantiate a `PlotTraining` object by providing the file paths for the epoch, batch, and validation log files.  
- This setup enables seamless visualization and monitoring of training metrics using the provided logs.

```python
from netweaver.utils import PlotTraining
```

```python
plot = PlotTraining(  # replace the params value with your training log paths
    path_epoch_log="/home/USER_NETWEAVER/project_netweaver/notebooks/demonstration/logs/model_training/model-20250523-132244/Log-epoch-20250523-132244.csv",
    path_batch_log="/home/USER_NETWEAVER/project_netweaver/notebooks/demonstration/logs/model_training/model-20250523-132244/Log-batch-20250523-132244.csv",
    path_validation_log="/home/USER_NETWEAVER/project_netweaver/notebooks/demonstration/logs/model_training/model-20250523-132244/Log-validation-20250523-132244.csv",
)
```

- The `live_plot` method reads the provided log files and dynamically updates a single graph for clear, real-time interpretation of training progress.

- The plot displays:
    - **x-axis:** Epochs
    - **Left y-axis (1):** Loss values
    - **Right y-axis (2):** Accuracy values
    - **Right y-axis (3):** Learning rate values
- This visualization enables easy monitoring of key training metrics as they evolve during the training process.

```python
plot.live_plot(loss_caplimit=2)
```

- The `pause()` method temporarily halts the live animation of the training metrics plot, allowing you to inspect the current state of the graph without further updates. This is useful for detailed analysis or presentations.

```python
plot.pause()
```

- The `resume()` method restarts the live animation of the training metrics plot, allowing real-time updates to continue after a pause. This is useful for resuming monitoring and visualization of training progress without losing context.


```python
plot.resume()
```

- Use the `static_plot()` method to generate a comprehensive visualization of all training metrics after training has completed.  

- This method is ideal for summarizing results and can also save the plot as an image file for reporting or further analysis.

```python
plot.static_plot()
```
